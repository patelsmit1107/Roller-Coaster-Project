```python
#change opening_date_clean to datetime
df['opening_date_clean']=pd.to_datetime(df['opening_date_clean'])
```

```python
df.dtypes #see data type of opening_date_clean is changed
```

```python
#rename columns
df=df.rename(columns={'coaster_name':'Coaster_Name',
            'year_introduced':'Year_Introduced',
            'opening_date_clean':'Opening_Date_Clean',
            'speed_mph' : 'Speed_mph',
             'height_ft': 'Height_ft',
              'Inversions_clean':'Inversions',
               'Gforce_clean':'Gforce' })
```

```python
#check null values
df.isna().sum()
```

```python
#cheking duplicate data
df.duplicated()
```

```python
#remove duplicate
df.columns
```

```python
df.duplicated(subset=['Coaster_Name','Location','Opening_Date_Clean']).sum() #tere will be 97 duplicated
```

```python
df.loc[~df.duplicated(subset=['Coaster_Name','Location','Opening_Date_Clean'])].head() #this are not duplicated columnswe gor by usinf(~)inverse of duplicated
```

```python
df=df.loc[~df.duplicated(subset=['Coaster_Name','Location','Opening_Date_Clean'])].reset_index(drop=True).copy() #changed index
```

What are the loction with the fastest roller costers(minimum 10 l0cation)?

```python
ax=df.query("Location != 'Other'" )\
  .groupby('Location')['Speed_mph']\
  .agg(['mean','count'])\
  .query('count >= 10')\
  .sort_values('mean')['mean']\
  .plot(kind='barh',title='Average Coast Speed By Location',color='green',figsize=(12,5))
plt.xlabel('Average Coaster Speed')
ax.grid(False)
plt.show()

  #query is used for filtering it will gives data where location is not equal to other for better analysis
  #we group by location and speed based on speed
  #aggregation function return mean,count of the coulumns
  #again we filter the data where count is more than 10
  #and we sort the values by mean
```
